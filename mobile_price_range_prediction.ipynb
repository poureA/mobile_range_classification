```python
import pandas as pd 
```

```python
file = pd.read_csv('F:\\datasets\\mobile_range_class\\train.csv')
df = file.copy()
df.sample(10)
```

```python
label = df['price_range']
df = df.drop(['price_range'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df = pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
print('scaling done !')
```

```python
from sklearn.model_selection import train_test_split
xtr , xts , ytr , yts = train_test_split(scaled_df,label,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
alg = RandomForestRegressor()
alg.fit(xtr,ytr)
alg.score(xtr,ytr)*100
```

```python
alg.score(xts,yts)*100
```
